# Load primary dataset

```python
from datasets import load_dataset

# Load primary dataset
print("Loading primary dataset...")
primary_dataset = load_dataset("sahil2801/CodeAlpaca-20k")
print(primary_dataset)
```

# Login with HF token

```python
from huggingface_hub import login
from google.colab import userdata

hf_token = userdata.get('HF_TOKEN')
login(token=hf_token)
print("Logged in successfully")
```

# Explore primary dataset

```python
import pandas as pd

# Convert to dataframe
df_primary = pd.DataFrame(primary_dataset["train"])

# Basic info
print(f"Dataset size: {df_primary.shape}")
print(f"Columns: {df_primary.columns.tolist()}")
print(f"\nFirst 2 rows:")
df_primary.head(2)
```

# Save Raw Primary Data In Locally

```python
import json

# Save primary raw dataset
with open("codealapaca_20k_raw.jsonl", "w") as f:
    for sample in primary_dataset["train"]:
        f.write(json.dumps(sample) + "\n")

print(f"Saved {len(primary_dataset['train'])} samples to codealapaca_20k_raw.jsonl")
```

# Save Raw Backup Data In Locally

```python
# Save backup raw dataset
with open("python_code_instructions_18k_raw.jsonl", "w") as f:
    for sample in backup_dataset["train"]:
        f.write(json.dumps(sample) + "\n")

print(f"Saved {len(backup_dataset['train'])} samples to python_code_instructions_18k_raw.jsonl")
```
