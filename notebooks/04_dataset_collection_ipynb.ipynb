# Load primary dataset

```python
from datasets import load_dataset

# Load primary dataset
print("Loading primary dataset...")
primary_dataset = load_dataset("sahil2801/CodeAlpaca-20k")
print(primary_dataset)
```

# Login with HF token

```python
from huggingface_hub import login
from google.colab import userdata

hf_token = userdata.get('HF_TOKEN')
login(token=hf_token)
print("Logged in successfully")
```

# Explore primary dataset

```python
import pandas as pd

# Convert to dataframe
df_primary = pd.DataFrame(primary_dataset["train"])

# Basic info
print(f"Dataset size: {len(df_primary)}")
print(f"Columns: {df_primary.columns.tolist()}")
print(f"\nFirst 3 rows:")
df_primary.head(3)
```

#  Preview single sample

```python
# Preview first sample in detail
sample = primary_dataset["train"][0]
print("INSTRUCTION:", sample["instruction"])
print("\nINPUT:", sample["input"])
print("\nOUTPUT:", sample["output"])
```

# Load backup dataset

```python
# Load backup dataset
print("Loading backup dataset...")
backup_dataset = load_dataset("iamtarun/python_code_instructions_18k_alpaca")
print(backup_dataset)
```

# Explore backup dataset

```python
# Convert to dataframe
df_backup = pd.DataFrame(backup_dataset["train"])

# Basic info
print(f"Dataset size: {len(df_backup)}")
print(f"Columns: {df_backup.columns.tolist()}")
print(f"\nFirst 3 rows:")
df_backup.head(3)
```

```python
# Compare both datasets
print("=" * 50)
print("DATASET COMPARISON")
print("=" * 50)
print(f"\nPrimary — CodeAlpaca-20K:")
print(f"  Samples : {len(df_primary)}")
print(f"  Columns : {df_primary.columns.tolist()}")
print(f"  Source  : sahil2801/CodeAlpaca-20k")
print(f"  License : Apache 2.0")

print(f"\nBackup — Python Code Instructions 18K:")
print(f"  Samples : {len(df_backup)}")
print(f"  Columns : {df_backup.columns.tolist()}")
print(f"  Source  : iamtarun/python_code_instructions_18k_alpaca")
print(f"  License : Apache 2.0")

print(f"\nTotal combined samples: {len(df_primary) + len(df_backup)}")
```
